```python
import numpy as np

class Module:
    
    def __init__(self):
        self.test_mod = {}

    def __setattr__(self, name, value):
        if isinstance(value, Module):
            self.test_mod[name] = value
        object.__setattr__(self, name, value)

    def __repr__(self):
        lines = [self.__class__.__name__ + "("]

        for name, module in self.test_mod.items():
            module_str = repr(module)
            lines.append(f"  ({name}): {module_str}")

        lines.append(")")
        return "\n".join(lines)
```

```python
test_module = Module()
test_module
```

```python
class Linear(Module):

    def __init__(self, in_features, out_features, bias=True):
        super().__init__()

        self.in_features = in_features
        self.out_features = out_features

        # initialize weights
        self.weight = np.random.randn(out_features, in_features) * 0.02

        if bias:
            self.bias = np.zeros(out_features)
        else:
            self.bias = None

    def __call__(self, x):

        y = x @ self.weight.T

        if self.bias is not None:
            y = y + self.bias

        return y

    def __repr__(self):
        return f"Linear(in_features={self.in_features}, out_features={self.out_features}, bias={self.bias is not None})"
```

```python
class Attention(Module):

    def __init__(self):

        super().__init__()

        self.q_proj = Linear(2048, 2048)
        self.k_proj = Linear(2048, 256)
        self.v_proj = Linear(2048, 256)
        self.o_proj = Linear(2048, 2048, bias=False)

    def forward(self, x):

        q = self.q_proj(x)
        k = self.k_proj(x)
        v = self.v_proj(x)

        return q, k, v
```
